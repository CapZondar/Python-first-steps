```python
import pandas as pd
```

```python
#Initial loading of data for the following exercise

df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data', header = None)
df
```

```python
#Renaming of columns for easier identification

df.columns=["Flow_ID","Date","Time","Count"]
df.head()

#Flow ID description:
#Flow 7 = Out
#Flow 9 = In
```

```python
#Filter presets

july24_2005 = df['Date'] == '07/24/05'
flow_out = df['Flow_ID'] == 7
```

```python
#Filtered data showing out flow for July 24, 2005

filtered_df = df.loc[july24_2005 & flow_out]
filtered_df.head()
```

```python
#Sorting count by descending order and assigning it as variable

top10 = filtered_df.sort_values(by="Count", ascending=False).head(10)

top10
```

```python
#Practicing pandas apply function

x = top10['Count'].min()
def sub(x):
  #for y in top10['Count']:
  #note to self: NEVER combine FOR LOOP and APPLY since both of them applies the function for each row
    return(x-1)
print(x)
```

```python
#Continuation of previous practice

z = top10['Count'].apply(sub)
z.mean()
```

```python
#Grouping the data for the month of August showing only flow id 7 (outflow)
df_august_1 = df['Date'] >= '07/31/2005'
df_august_2 = df['Date'] <= '08/31/2005'

august_flow_out = df[df_august_2 & df_august_1 & flow_out]

august_flow_out.groupby('Date').max('Count')
```
